# Linear Regression Using Tensorflow

# Import libaries 

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

# Design our model using the Sequential, which is a linear stack of layers. In this model, we use only one layer(neuron).

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

# we define our optimizer and loss functions to train our Neural Network model. In this, we use the Stochastic Gradient Descent (SDG) optimizer and the Mean Squared Error (MSE) as our loss function.

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

# we initialize our numbers with two variables “xs” and “ys”.

```python
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

# we fit our model to the variables “xs” and “ys”. We train the model with 500 epochs.

```python
model.fit(xs, ys, epochs=500)
```

# we predict the value for a new xs = 10. According to the formula, y=2*x-1, for xs=10, we get a value of 19.

```python
print(model.predict([10.0]))
```

```python
model.save("Linear regression using Tensorflow.h5")
```
